```python
import numpy as np
```

```python
#calculate California daily data
daily_data_PSCA_data=[]
for i in range(len(train)):
    first = np.sum(np.array(list(train.values()))[i,0:24,:],axis=0)
    for j in range(1,365):
        first = np.vstack([first,np.sum(np.array(list(train.values()))[i,24*j:24*(j+1),:],axis=0)])
    daily_data_PSCA_data.append(first)
daily_data_PSCA_data=np.array(daily_data_PSCA_data)
```
